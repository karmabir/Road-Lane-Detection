```python
import matplotlib.pylab as  plt
import cv2
import numpy as np
```

```python
#Mask everyother thing except ROI
def roi(image, vertices):
    mask = np.zeros_like(image)
    #channel_count = img.shape[2]
    match_mask_color = 255
    cv2.fillPoly(mask, vertices, match_mask_color)
    masked_image = cv2.bitwise_and(image, mask)
    return masked_image

#To draw lines
def drawlines(image, lines):
    copy_image = np.copy(image)
    line_image = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
    
    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(line_image, (x1, y1), (x2,y2), (255,0,0), thickness=3)
            
    image = cv2.addWeighted(image, 0.8, line_image, 1, 0.0)
    return image

#img = cv2.imread('D:\Projects\OpenCV\Image\Road3.jpg')
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def process(img):

    print(img.shape)
    height = img.shape[0]
    width = img.shape[1]

    #ROI
    roi_vertices = [
        (0, height),
        (width/2, height/2),
        (width, 350),
        (width, height)
    ]

    gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    canny_img = cv2.Canny(gray_img, 100,120)
    cropped = roi(canny_img, np.array([roi_vertices], np.int32))

    lines = cv2.HoughLinesP(cropped, 
                           rho = 6,
                           theta = np.pi/180,
                           threshold = 160,
                           lines = np.array([]),
                           minLineLength=40,
                           maxLineGap=100)

    imagelines = drawlines(img, lines)
    return imagelines

cap = cv2.VideoCapture('test2.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    frame = process(frame)
    
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(40) == 27:
        break   
cap.release()
cv2.destroyAllWindows()
```
